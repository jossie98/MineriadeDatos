```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from apyori import apriori ##Esta es la libreria que usaremos para realizar la clasificacion asocierada
```

```python
#Importamos la BD
path ='base.csv'
df = pd.read_csv(path)
df.head()
```

```python
df.shape  #Nos premitira conocer el numero de columnas y de registros de las compras
```

```python
records=[]
for i in range(0,7500):
    records.append([str(df.values[i,j]) for j in range(0,20)]) 
```

```python
records #Los ponemos en una lista
```

```python
association_rules = apriori(records, min_support=.003050,min_confidence=0.2,min_lift=3,min_length=2)   #30/9835 = 0.003050
```

```python
association_results=list(association_rules) 
```

```python
for element in association_results:
    print (element)
```
