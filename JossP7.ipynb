```python

import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

#Aqui cargamos los datos 
dataset = pd.read_csv('diabetes.csv')
```

```python
dataset
```

```python
#Elimino las columnas no necesarias, para esta practica usaré las columnas Glucoese y BloodPressure
dataset.drop(['Pregnancies', 'SkinThickness','Insulin','DiabetesPedigreeFunction','Age','Outcome','BMI'], axis=1,inplace=True) #Eliminamos las columnas que no utilizaremos
```

```python
#El data set queda de la siguiente manera
dataset.head()
```

```python
# Dividimos las var. dependientes e indepentiendes
#Glucosa y Salario
x = dataset.iloc[:, :-1].values
y = dataset.iloc[:, 1].values
```

```python
x
```

```python
y
```

```python
plt.scatter(x, y)
plt.xlabel('Glucosa')
plt.ylabel('Presion Sanguinea')
plt.show()
#Aqui se muestra la dispresion de los datos
```

```python
#Hcemos uso de la sifuiente libreria
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.2)
```

```python
from sklearn import datasets, linear_model
lr = linear_model.LinearRegression()
```

```python
lr.fit(X_train, y_train)
```

```python
Y_pred = lr.predict(X_test) #Prediccion
```

```python

plt.scatter(X_test, y_test)
plt.plot(X_test, Y_pred, color='green', linewidth=4)
plt.ylabel('Glucosa')
plt.xlabel('Presion')
plt.show()
```
